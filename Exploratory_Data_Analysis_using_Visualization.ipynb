```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Basic exploration
print(df.info())
print(df.describe())
```

```python
#Data Cleaning

df['age'] = df['age'].fillna(df['age'].median())
df.drop(columns=['deck'], inplace=True)
```

```python
#Q1 :- Passenger Count by Gender

plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=df, palette='Set2')
plt.title('Number of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Q2 :- Passenger Count by Class

plt.figure(figsize=(6,4))
sns.countplot(x='pclass', data=df, palette='viridis')
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
#Q3 :- Distribution of Passenger Age

plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True, color='blue')
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q4 :- Fare Outlier Detection

plt.figure(figsize=(6,4))
sns.boxplot(x=df['fare'], color='salmon')
plt.title('Boxplot of Fare to Detect Outliers')
plt.xlabel('Fare')
plt.show()
```

```python
#Q5 :- Scatterplots: Age, Fare, and Family Size

# a. Relationship between Age and Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='fare', data=df, alpha=0.5)
plt.title('Relationship between Age and Fare')
plt.show()

# b. Family Size (SibSp + Parch) and Fare
df['family_size'] = df['sibsp'] + df['parch']
plt.figure(figsize=(8,5))
sns.scatterplot(x='family_size', y='fare', data=df, color='green')
plt.title('Relationship between Family Size and Fare')
plt.show()
```

```python
#Q6 :- Survival Count by Gender

plt.figure(figsize=(6,4))
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival Count by Gender')
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```

```python
#Q7 :- Survival by Passenger Class

plt.figure(figsize=(6,4))
sns.barplot(x='pclass', y='survived', data=df, ci=None, palette='plasma')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Q8 :- Correlation Heatmap

plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
